```python
def etiquetar_edad(edad:int)->bool:
  """" Retorna una etiqueta según mayoría de edad """
  if edad>=15:
    return True
  else:
    return False
```

```python
etiquetar_edad(20)
```

```python
df['es_mayor_edad'] = df.Age.apply(etiquetar_edad)
```

```python
# Funcion De Guardado
def guardar_df(df: pd.DataFrame,sheet_name:str, file_name:str, folder:str ):
    
    if not os.path.isdir(folder):
        os.mkdir(folder)
    
    path = f'{folder}/{file_name}.xlsx'
    df.to_excel(path, sheet_name=sheet_name, index=False)
    pass
```

```python
guardar_df(df,sheet_name='titanic', file_name='titanic_procesado', folder='./out' )
```
