Extra: Colocar una marca para las personas mayores de edad. Para ello debemos generar una nueva columna

```python
df.head()
```

```python
def es_mayor_edad(edad :int) -> str:
  if edad>18:
    return "Mayor Edad"
  else:
    return "Menor Edad"

```

```python
es_mayor_edad(21)
```

```python
df['mayor_edad'] = df.Age.apply(es_mayor_edad)
```

#### Iniciando con los gráficos

```python
%matplotlib inline
```

1. Diagrama de sectores(circular) con los fallecidos y supervivientes.

```python
df.Survived.value_counts().plot.pie(
    labels=["Fallecidos", "Sobrevivientes"],
    colors=["r", "g"],
    autopct="%.2f",
    fontsize=20,
    figsize=(6, 6)

);
```

2. Diagrama de barras con el número de personas en cada clase

```python
df.groupby('Pclass')['PassengerId'].count()
```

```python
df.Pclass.value_counts()
```

```python
df.Pclass.value_counts().plot.bar()
```

4. Diagrama de barras con el número de personas fallecidas y supervivientes en cada clase.
